```python
columns = [
    "departure_city", "arrival_city",
    "company_name", "fare_class_code"
]
```

```python
encoder = ce.BinaryEncoder(cols=columns)
encoded_data = encoder.fit_transform(df)
```

```python
encoded_data.head()
```

```python
one_hot_encoded = pd.get_dummies(encoded_data, columns=["flight_sale_type", "dep_flight_time"])
```
